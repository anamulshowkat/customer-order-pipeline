```python


# File path define
file_path = "/Workspace/Users/anaamfaisal@gmail.com/customer-order-pipeline/volumes/customers raw.csv"

# CSV read 
df_bronze = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)
# Bronze table path save (Delta format)
df_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.my_database.customers_bronze")

print("Bronze Layer loaded successfully!")
display(df_bronze)

```
