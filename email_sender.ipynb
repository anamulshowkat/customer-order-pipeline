```python
import smtplib
from email.mime.text import MIMEText

def send_view_report():
    # 1. Fetch Data
    df = spark.table("workspace.my_database.vw_customers_orders_master")
    total_rows = df.count()
    total_customers = df.select("customer_id").distinct().count()
    total_sales = df.agg({"total_amount": "sum"}).collect()[0][0]
    
    # 2. Content
    report_body = f"Hello Team,\n\nThe Master View has been updated.\n- Total Records: {total_rows}\n- Unique Customers: {total_customers}\n- Total Sales: {total_sales}"
    
    # 3. SMTP Server Details (Example for Gmail)
    smtp_server = "smtp.gmail.com"
    smtp_port = 587
    sender_email = "anaamfaisal@gmail.com"
    # IMPORTANT: IF you are using Gmail THEN  use 'App Password' instead of your actual password Password
    password = "ckmt awom nsls xiou" 
    receiver_email = "anaamshah9999@gmail.com"
    
    # 4. Create and Send Email
    msg = MIMEText(report_body)
    msg['Subject'] = "Master View Data Report"
    msg['From'] = sender_email
    msg['To'] = receiver_email
    
    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls() # Secure the connection
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, msg.as_string())
        server.quit()
        print("Email sent successfully!")
    except Exception as e:
        print(f"Error sending email: {e}")

# Run the function
send_view_report()
```
